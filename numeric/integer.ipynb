```python
# Python uses dynamic size to store numbers
# that means the memory block size that store the integer object grow if the number get very large
# this is very different than Java, C, etc 
```

```python
import sys
```

```python
sys.getsizeof(100)
```

```python
a = 2**1000000
a
```

```python
sys.getsizeof(a)
```

```python
# so we can see that Python can handle a super large number
```
